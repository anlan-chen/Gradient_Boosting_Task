```python
from sklearn.metrics import mean_squared_error, make_scorer
import numpy as np

# sklearn  RMSE scorer neg_root_mean_squared_error 
rmse_scorer = "neg_root_mean_squared_error"

#  RMSE 
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

def tune_model_with_grid(
    estimator,
    param_grid,
    X,
    y,
    model_name="",
    feature_set_name="top50",
    n_splits=5,
):
    """
     +  GridSearchCV CV
    
      - best_estimator_
      - cv_results_ DataFrame
    """
    print(f"=== Tuning {model_name} on feature set: {feature_set_name} ===")
    print("Param grid:", param_grid)

    tscv = TimeSeriesSplit(n_splits=n_splits)

    gs = GridSearchCV(
        estimator=estimator,
        param_grid=param_grid,
        cv=tscv,
        scoring=rmse_scorer,     #  neg_root_mean_squared_error
        n_jobs=-1,               # 
        verbose=2,
    )

    gs.fit(X, y)

    print(f"[{model_name}] Best params: {gs.best_params_}")
    print(f"[{model_name}] Best CV RMSE: {-gs.best_score_:.4f}")

    cv_results = (
        pd.DataFrame(gs.cv_results_)
        .sort_values("rank_test_score")
        .reset_index(drop=True)
    )

    return gs.best_estimator_, cv_results
```
